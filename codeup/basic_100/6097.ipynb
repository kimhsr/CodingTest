```python
game_board = list()

h, w = map(int, input().split())

# 2차원 리스트 초기화
for i in range(h):
  tmp = []
  for j in range(w):
    tmp.append(0)
  game_board.append(tmp)

n = int(input())
for i in range(n):
  l, d, x, y = map(int, input().split())
  x -= 1
  y -= 1

  if d == 0: # 가로로 놓인 경우
    for i in range(l):
      game_board[x][y+i] = 1
  elif d == 1: # 세로로 놓인 경우
    for i in range(l):
      game_board[x+i][y] = 1

# 결과 출력
for i in range(h):
  for j in range(w):
    print(game_board[i][j], end=' ')
  print()
```
