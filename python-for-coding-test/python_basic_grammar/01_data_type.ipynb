```python
# 양의 정수
a = 1000
print(a)

# 음의 정수
a = -7
print(a)

# 0
a = 0
print(a)
```

```python
a = 777
print(a)

a = a + 5
print(a)

a = a - 3
print(a)
```

```python
# 양의 실수
a = 157.93
print(a)

# 음의 실수
a = -1837.2
print(a)

# 소수부가 0일 때 0을 생략
a = 5.
print(a)

# 정수부가 0일 때 0을 생략
a = -.7
print(a)
```

```python
# 1,000,000,000의 지수 표현 방식
a = 1e9
print(a)

# 752.5
a = 75.25e1
print(a)

# 3.954
a = 3954e-3
print(a)
```

```python
# 실수형을 정수로 바꾸기 int
a = int(1e9)
print(a)
```

```python
a = 0.3 + 0.6
print(a)

if a == 0.9:
  print(True)
else:
  print(False)
```

```python
a = 0.3 + 0.6
print(round(a, 1))

if round(a, 1) == 0.9:
  print(True)
else:
  print(False)
```

```python
a = 7
b = 3

# 나누기
print(a / b)

# 나머지
print(a % b)

# 몫
print(a // b)
```

```python
a = 5
b = 3

# 거듭 제곱
print(a ** b)

# 제곱근 
print(a ** 0.5)
```
